# 4.데이터분석을 해봅시다

데이터 분석은 복잡한 수식이 들어간 멋진 모델링에서 시작되는 것도 아니고, 화려한 시각화에서 시작되는 것도 아니다. 시각화와 모델링을 할 수 있도록 데이터를 준비하는 과정, 즉 전처리를 하면서 데이터를 이해하는 작업이 필요하다. 지루하고 따분한 과정이라고 생각할 수 있지만 데이터를 꼼꼼히 들여다보며 각각의 변수가 어떤 의미인지, 어떤 방식으로 측정된 것인지, 명목형인지 수치형인지,범주형인지 등 실제 의미대로 코딩되었는지 정확히 이해하는 이 과정은 데이터로부터 좋은 인사이트를 얻는 출발점이라고 할 수 있다.

물론 직접 데이터를 수집했다면 각각의 변수가 무엇을 나타내는지, 수치형 혹은 문자형으로 표현된 관측값이 무엇을 의미하는지 이미 알고 있을 수 있다. (그래도 시간이 지나면 거짓말처럼 내가 뭘 주워담은 건지 잊어버리게 되니 처음 데이터를 수집했을 때 데이터 설명서를 잘 써둬야 한다.) 하지만 다른 사람이 수집한 데이터를 분석해야 할 때는 어떻게 해야 할까? 예측모델 및 분석 대회 플랫폼으로 유명한 캐글(Kaggle)에 가보면 다양한 주제의 데이터를 구할 수 있다. 여기서 얻은 데이터로 모델링이나 시각화를 해보기 전, 가장 먼저 할 일은 바로 ‘데이터 설명서(data description) 읽기’다.

## 2)EDA(Exploratory Data Analysis,탐색적 데이터 분석)

EDA(Exploratory Data Analysis,탐색적 데이터 분석) : 

본격적인 모델링에 들어가기에 앞서, 반드시 선행되어야 하는 과정

데이터의 분포나 변수간 관계를 파악하기 위해 히스토그램, 산점도, 상관관계표 등 다양한 시각화 방법이 동원된다.


데이터 전처리에 앞선 데이터의 종류

​

유형 1. 수치형으로 입력되어 있지만 실체는 명목형인 변수

유형 2. 명목형으로 입력되어 있지만 실제는 순서상의 의미를 가지는 변수

유형 3. 합쳐서 하나로 만들 수 있는 변수

유형 4. 쪼개서 나눌수 있는 변수

유형 5. 결측값인지, 값이 0인지 헷갈리는 관측치 등

​

81개나 되는 변수가 있지만 본격적인 EDA 및 모델링에 들어가기에 앞서 변수 하나 하나를 살펴보며 제대로 인코딩해나가는 과정이 있어야 실제에 가까운, 더 많은 정보를 담은 데이터가 준비된다. 이렇게 변수의 유형을 나눠가며 인코딩하다보면 금방 데이터의 특성을 파악할 수 있을 것이다.

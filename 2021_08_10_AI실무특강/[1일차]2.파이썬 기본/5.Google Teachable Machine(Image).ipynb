## 지도학습 (Supervised Learning)
- 데이터의 레이블 (또는 클래스) 정보를 가지고 학습하는 방법으로 모델에는 분류(Classification)와 예측(Prediction)이 있음

## 분류 (Classification)
- 레이블이 있는 데이터를 가지고 학습한 후 새로운 데이터를 입력했을 때, 그 데이터의 레이블을 판별
- 예: 0 ~ 9 까지의 숫자 이미지 데이터가 있고, 각 숫자가 데이터의 레이블일 때, 이 훈련데이터 0 ~ 9 사이의 숫자를 판별하는 분류 모델을 학습하여 구축. 이 후, 새로운 숫자 이미지 데이터를 분류 모델에 집어 넣고 그 숫자의 이미지가 어떤 레이블인지 판별
- Image, Audio 등을 이진분류 (Binary Classification)하거나 다중분류 (Multiclass Classification) 함 

## Classification 절차 
- 레이블 가진 훈련 데이터 수집
- 훈련(Training)
- 테스트 데이터 입력 후 분류 모델 구축
- 테스트 데이터의 레이블 판별

## Classification 핵심 용어
- 레이블(Label):  데이터가 속한 그룹의 이름(개/고양이 등)
- 훈련(Training): 주어진 데이터를 이용한 기계 학습 활동
- 모델(Model): 학습을 통해 기계가 얻은 결과물
- 테스트(Testing) 또는 추론(Inference): 구축된 모델이 정확한지 시험

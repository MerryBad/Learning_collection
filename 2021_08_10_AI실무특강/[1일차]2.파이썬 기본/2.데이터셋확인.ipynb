```python
for i in range(0, len(d.data)):
    print(i,d.data[i], d.target[i])
```

```python
from sklearn import svm
```

```python
s = svm.SVC(gamma=0.1, C=10)

s.fit(d.data, d.target)
```

```python
new_d = [[6.4, 3.2, 6.0, 2.5],[7.1,3.1,4.7,1.35],[7.1,3.1,4.7,1.35]]
```

```python
res = s.predict(new_d)
```

```python
print("새로운 2개 샘플 분류", res)
```

# 싸이킷런에서 당뇨병 데이터셋 적재

```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import datasets
# 당뇨병 데이터 셋을 잭재한다
daiabetes = datasets.load_diabetes()
```

# 데이터셋 열수와 컬럼수 확인

```python
daiabetes.data.shape
```

# 데이터셋 컬럼명 확인

```python
daiabetes.feature_names
```

# 데이터셋 설명서 

```python

print(daiabetes.DESCR)
```

# 내부 데이터 확인(10가지 컬럼 데이터와 1개의 정답(1년동안의 당수치 변화,라벨)으로 구성)

```python
for i in range(0, len(daiabetes.data)):
    print(i,daiabetes.data[i], daiabetes.target[i])
```

# Train 데이터와 Test 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(daiabetes.data, daiabetes.target, test_size=0.2, random_state=0)
```

# 선형회기 모델로 학습(model.fit)

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

# 테스트 데이터로 예측

```python
# 테스트 데이터로 예측해보자
y_pred = model.predict(X_test)
```

# 실제 데이터와 예측 데이터 비교

```python
# 실제 데이터와 예측 데이터를 비교해보자
plt.plot(y_test,y_pred, '.')
```

# 점선과 그래프 겹쳐서 보기(선형성 확인)

```python
plt.plot(y_test,y_pred, '.')
x = np.linspace(0,330,100)
y = x
plt.plot(x,y)
plt.show()
```

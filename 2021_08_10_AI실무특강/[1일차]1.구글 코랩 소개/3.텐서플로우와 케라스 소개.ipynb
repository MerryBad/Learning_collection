텐서는 배열을 담는 컨테이너(공간)

텐서플로우는 딥러닝 프레임워크의 일종이다. 텐서플로우는 내부적으로 C/C++로 구현되어 있고 파이썬을 비롯하여 여러가지 언어에서 접근할 수 있도록 인터페이스를 제공한다. 텐서플로우는 텐서(Tensor)와 플로우(Flow)가 합쳐진 용어 이다. 텐서는 물리학에서 다차원 배열을 나타내는 용어이다. 배열들을 담는 컨테이너라고 생각하는 것이 좀더 도움이 된다. 

구글 코랩에서는 텐서 플로우가 기본적으로 설치되어 있다. 

딥러닝에서 텐서란...
* 다차원 배열을 텐서라고 부름
* 텐서는 다차원 배열을 담는 컨테이너라고 인식하는 것이 좋음
* 신경망의 가중치(매개변수)를 텐서로 표현

```python
import tensorflow as tf

print(tf.__version__)
a=tf.random.uniform([2,3],0,1) #  3차원 벡터의 배열 2개를 담는 텐서 생성, 데이터는 0~1사이 랜덤값
print(a)
print(type(a))
```

```python
import tensorflow as tf
import numpy as np

t=tf.random.uniform([2,3],0,1)  # 서로 함수 파라미터의 사용법이 다름
n=np.random.uniform(0,1,[2,3])  # 서로 함수 파라미터의 사용법이 다름

print("텐서플로우로 생성한 텐서:\n",t,"\n")
print("넘파이로 생성한 텐서:\n",n,"\n")

res=t+n
print("텐서간 덧셈 결과:\n",res)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 성적 분석 예제
```

```python
df = pd.read_csv("data/score.csv")
df
```

```python
df.info()
```

```python
df.describe()  # 통계 정보
```

```python
subjects = [ '국어', '영어', '수학', '과학' ]  # 리스트 : 배열
```

```python
# df['추가할_컬럼'] = 계산식
df['총점'] = df[ subjects ].sum( axis=1 )  # axis=0 디폴트:행, axis=1 : 열
df
```

```python
df['평균'] = df[subjects].mean(axis=1)
df
```

```python
# 정렬 : 평균이 높은 학생부터 낮은 학생 순(내림차순), 평균이 같으면 이름은 오름차순으로 출력
sorted_df = df.sort_values( ['평균', '이름'], ascending=[False, True] )
sorted_df
```

```python
# 그래프를 그릴 때 한글깨짐 방지
# 원인 : 디폴트 글꼴이 영어이기 때문
# 해결 : 한글글꼴로 변경
import matplotlib
matplotlib.rc('font', family='Malgun Gothic')
```

```python
# 시각화
sorted_df['평균'].plot(kind='bar')
```

```python
sorted_df.index = sorted_df['이름']
sorted_df
```

```python
sorted_df['평균'].plot(kind='bar', figsize=(12, 8))  # (12, 8) <== tutple
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/titanic.xls")
df
## feature selection : 특성 선택
##### X
# pclass : 선실등급(1,2,3), 1등실 가장 좋다, 범주형(순서형), object
# sex : 성별(female:여성, male:남성), 범주형(명목형), object
# age : 나이, 수치형(연속형), float64
# sibsp : 부부형제자매_수, 수치형(이산형), int64
# parch : 부모자식_수, 수치형(이산형), int64
# fare : 운임, 수치형(연속형), float64
##### y 정답 : survived(0:사망, 1:생존), 이진분류
```

```python
df.info()
```

```python
2 ** 63  # 922경(int64), int32는 플러스 마이너스 21억
```

```python
df.describe()  # 통계 정보
```

```python
# 결측치 갯수 출력
df.isnull().sum()  # axis=0 이 디폴트
```

```python
# 운임이 결측치인 사람을 출력하시오
# df.get( df['fare'].isnull() )
df[ df['fare'].isnull() ]  # df[ 조건 ] 
```

```python
df.describe()
```

```python
df.boxplot()
```

```python
# 생존자의 수와 사망자 수를 출력하시오
print("생존자 수:", df['survived'].sum() )  # 생존자 수
print("사망자 수:", 1309 - df['survived'].sum() )    # 사망자 수
print("사망자 수:", df['survived'].count() - df['survived'].sum() ) # 사망자 수
```

```python
# 생존자의 수와 사망자 수를 출력하시오
df['survived'].value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')  # python warning 무시
```

```python
# 시각화
plt.figure(figsize=(12,6))
plt.subplot(121)
df['survived'].value_counts().plot.pie(explode=[0, 0.1], autopct="%.2f%%")
plt.subplot(122)
sns.countplot('survived', data=df)
plt.show()
```

```python
# 시각화 : 플러스 알파 (countplot 에 값들을 출력)
plt.figure(figsize=(12,6))
plt.subplot(121)
df['survived'].value_counts().plot.pie(explode=[0, 0.1], autopct="%.2f%%")
plt.subplot(122)
g1 = sns.countplot('survived', data=df)
for p in g1.patches:
    g1.annotate( f"{p.get_height()}", (p.get_x()+0.3, p.get_height()+5) )
plt.show()
```

```python
# 연령대 그래프 : 분포도 (히스토그램)
df['age'].hist(bins=20, grid=False, figsize=(18, 8))
```

```python
# 성별과 생존의 관계 출력
df.groupby(['sex', 'survived'])['survived'].count()
```

```python
df[ ['sex', 'survived'] ].groupby(['sex']).mean()
```

```python
# 남여별 생존 그래프
plt.figure(figsize=(18, 8))
sub1 = plt.subplot(121)
df[ ['sex', 'survived'] ].groupby(['sex']).mean().plot.bar(ax=sub1)
plt.subplot(122)
sns.countplot('sex', hue='survived', data=df)  # 남여별 생존별 그래프
plt.show()
```

```python
# 선실 등급별 데이터 체크, 그래프
pd.crosstab(df['pclass'], df['survived'])
```

```python
pd.crosstab(df['pclass'], df['survived'], margins=True)
```

```python
# pd.crosstab(df['pclass'], df['survived']).plot(kind='bar')
pd.crosstab(df['pclass'], df['survived']).plot.bar()
```

```python
# 선실 등급별 통계(평균)
df.groupby('pclass').mean()
```

```python
# 상관 관계(상관 계수) 출력
df.corr()
```
